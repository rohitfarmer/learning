```python
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.preprocessing import RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel("OTU.xlsx", index_col=0, header=1)
```

```python
df_no_target = df.drop(labels="taxonomy", axis=1)
```

```python
scaler = RobustScaler()
df_trans = scaler.fit_transform(df_no_target)
```

```python
clf = LinearDiscriminantAnalysis()
```

```python
lda = clf.fit_transform(df_trans, df['taxonomy'])
```

```python
df_lda = pd.DataFrame(data=lda, index=df.index)
```

```python
ax = plt.figure(figsize=[10,10])
ax = sns.scatterplot(df_lda[0], df_lda[1], hue=df['taxonomy'])
ax = plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
